```python
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as skl
import numpy as np
import seaborn as sns
import folium
path = "diabetes.csv"
df = pd.read_csv(path)
df = df[['dma','2017+diabetes']]
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
newStates = [] # make the old not working to new
done = False
for i in df['dma']:
    done = False
    for x in states:
        if x in i and not done:
            newStates.append(x)
            done = True
newStates = pd.Series(newStates)
df['dma'] = newStates
print(df['dma'])
            
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
state_geo = state_geo = f'{url}/us-states.json'
frequency = df['dma'].value_counts().to_dict()
print(frequency)
flist = list(frequency.keys())
fvalues = list(frequency.values())
df['dma'] = pd.Series(flist)
#df['2017+diabetes'] = pd.Series(fvalues)
m = folium.Map(location=[39.0119, -98.4842], zoom_start=5)

folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=df,
    columns=['dma','2017+diabetes'],
    key_on='feature.id',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Diabetes Disease Rate'
).add_to(m)

folium.LayerControl().add_to(m)
m.save('mapChoro.html')



```
